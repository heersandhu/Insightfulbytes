```
#Compare Make and eletricity consumption per hour in city
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
df = pd.read_csv('battery vehicles.csv')
plt.figure(figsize=(10, 6))
sns.barplot(x='Make', y='City (kWh/100 km)', data=df, estimator='mean')
plt.title('Electricity consumption in city')
plt.xlabel('Make')
plt.ylabel('City (kWh/100 km)')
plt.xticks(rotation=45)
plt.show()
```

```
#Compare Make and eletricity consumption per hour on highway
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
df = pd.read_csv('battery vehicles.csv')
plt.figure(figsize=(10, 6))
sns.barplot(x='Make', y='Highway (kWh/100 km)', data=df, estimator='mean')
plt.title('Electricity consumption on highway')
plt.xlabel('Make')
plt.ylabel('Highway (kWh/100 km)')
plt.xticks(rotation=45)
plt.show()
```

```
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 8))
sns.scatterplot(x='Recharge time (h)', y='Range (km)', hue='Make', data=df, palette='tab10', alpha=0.7)
plt.title('Recharge Time vs. Range by Vehicle Make')
plt.xlabel('Recharge Time (hours)')
plt.ylabel('Range (km)')
plt.legend(title='Make', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```
